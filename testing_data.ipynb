```python
import sqlite3
conn = sqlite3.connect("data.sqlite")
cursor = conn.cursor()
```

```python
# Count fires within ~20km of Lake Tahoe center (39.09, -120.03)
cursor.execute("""
    SELECT COUNT(*) FROM Fires 
    WHERE LATITUDE BETWEEN 38.85 AND 39.35 
    AND LONGITUDE BETWEEN -120.25 AND -119.85
""")
print(f"Fires near Lake Tahoe: {cursor.fetchone()[0]}")
```

```python
cursor.execute("""
    SELECT NWCG_GENERAL_CAUSE, COUNT(*) as n, 
           AVG(FIRE_SIZE) as avg_size
    FROM Fires 
    WHERE LATITUDE BETWEEN 38.85 AND 39.35 
    AND LONGITUDE BETWEEN -120.25 AND -119.85
    GROUP BY NWCG_GENERAL_CAUSE
    ORDER BY n DESC
""")
for row in cursor.fetchall():
    print(row)
```
